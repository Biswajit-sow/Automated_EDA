# know About pandas profiling now ydata profiling -"https://pypi.org/project/ydata-profiling/"

```python
import seaborn as sns
df = sns.load_dataset('titanic')
df.head()
```

```python
from ydata_profiling import ProfileReport
profile = ProfileReport(df,explorative=True,html={'style': {'dark_mode': True}},title="Titanic data Analysis")#explorative =True bcz more and more information and uyou can use other parameters
profile.to_file(output_file='Titanic_data_analysis.html')
```

```python
import pandas as pd

# URL for the Telco Customer Churn dataset
url = 'https://raw.githubusercontent.com/IBM/telco-customer-churn-on-icp4d/master/data/Telco-Customer-Churn.csv'

# Read the CSV file directly into a DataFrame
df_churn = pd.read_csv(url)

# Now you can use it with ydata-profiling
df_churn.head()
```

```python
from ydata_profiling import ProfileReport
profile = ProfileReport(df_churn,explorative=True,html={'style': {'theme': 'cosmo'}},sort='ascending',title="Telco data Analysis")#explorative =True bcz more and more information and uyou can use other parameters
profile.to_file(output_file='Telco_customer_analysis.html')
```
